### 🖼️ 3. Computer Vision Preprocessing
Let's test the OpenCV image processing function. It resizes images to standard dimensions and applies an unsharp mask to improve the contrast of handwritten text, preparing the document for optimal AI readability.

```python
from utils.preprocessing import preprocess_image

# Look for a dummy or sample image to test, or see the preprocessing code structure
dummy_image = np.ones((500, 500, 3), dtype=np.uint8) * 240
# Add some fake text/shapes
cv2.putText(dummy_image, "MedReceipt Preprocessing Test", (30, 250), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (50, 50, 50), 2)

processed = preprocess_image(dummy_image)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(dummy_image)
axes[0].set_title("Original Image (Mockup)")
axes[0].axis("off")

axes[1].imshow(processed)
axes[1].set_title("Preprocessed Image (Mockup)")
axes[1].axis("off")
plt.show()
print("CV preprocessing pipeline ran successfully.")
```
